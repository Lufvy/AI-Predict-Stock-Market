# Data set

```python
import yfinance as yf
import pandas as pd

ihsg = "^JKSE"

tickers = [
    "BBCA.JK", "BBRI.JK", "BMRI.JK", "TLKM.JK", "ASII.JK",
    "BBNI.JK", "UNVR.JK", "ICBP.JK", "INDF.JK", "CPIN.JK",
    "PTBA.JK", "ANTM.JK", "ADRO.JK", "PGAS.JK", "TINS.JK",
    "MDKA.JK", "KLBF.JK", "AKRA.JK", "BRPT.JK", "EXCL.JK",
    "ERAA.JK", "JPFA.JK", "MYOR.JK", "SMGR.JK", "INTP.JK",
    "WSKT.JK", "WIKA.JK", "PTPP.JK", "SCMA.JK", "UNTR.JK",
    "HMSP.JK", "GGRM.JK", "INDY.JK", "TOWER.JK", "MIKA.JK",
    "ARTO.JK", "ELSA.JK", "INDF.JK", "ISAT.JK", "ITMG.JK",
    "SIDO.JK", "ADMR.JK", "BUKA.JK", "MDLN.JK", "BFIN.JK"
]

startDate = "2018-01-01"
endDate = "2025-01-01"

all_data = []

# Add IHSG
IHSGdata = yf.download(ihsg, start=startDate, end=endDate)
IHSGdata["Ticker"] = "^JKSE"
if isinstance(IHSGdata.columns, pd.MultiIndex):
    IHSGdata.columns = IHSGdata.columns.get_level_values(0)
IHSGdata.reset_index(inplace=True)
all_data.append(IHSGdata)

# Add tickers (LQ45)
for ticker in tickers:
    ticker_data = yf.download(ticker, start=startDate, end=endDate)

    if ticker_data["Open"].all().empty:
        print(f"No useful data for {ticker}, skipping.")
        continue

    ticker_data["Ticker"] = ticker
    
    if isinstance(ticker_data.columns, pd.MultiIndex):
        ticker_data.columns = ticker_data.columns.get_level_values(0)
    ticker_data.reset_index(inplace=True)
    all_data.append(ticker_data)

combined_df = pd.concat(all_data)

if "Adj Close" in combined_df.columns:
    combined_df.drop(columns=["Adj Close"], inplace=True)

combined_df.to_csv("Dataset.csv", index=False)
```
